# Rappels sur les Algorithmes de tri vus en Première

Pour commencer, on va importer la fonction `randint` du module `random` pour créer des tableaux contenant des nombres tirés au hasard.

```python
from random import randint
```

# Algorithme de tri par insertion

Le principe : 
* le tableau est décomposé en deux parties, à gauche, une partie déjà trié, à droite une partie non triée
* on prend la première carte non triée et on l'insère dans la partie des cartes triées, à sa place

```python
def tri_insertion(T, verbose=False):
    '''
    Entrées :
        T est un tableau d'entiers
        verbose par défaut vaut False. Si verbose vaut True, on affiche toutes les étapes
    Sortie : le tableau T avec les nombres triés dans l'ordre croissant
    '''
    
    # Avant de commencer le tri du tableau
    if verbose== True:
            print(T)
    
    # On commence le tri
    for i in range(1,len(T)) :
        # on prend un nouvel indice
        j = i
        while (j > 0) and (T[j-1] > T[j]) :
            # tant qu'à gauche, il y a un élément plus grand
            # on les échange
            T[j], T[j-1] = T[j-1], T[j]
            j = j - 1
            if verbose== True:
                print(T)

    return T
```

```python
# taille du tableau
n = 20

# création par compréhension
T=[randint (0 ,100) for i in range(n)]

tri_insertion(T, True)
```

# Algorithme de tri par sélection

Le principe : 
* le tableau est décomposé en deux parties, à gauche, une partie déjà trié, à droite une partie non triée
* on prend dans la partie non triée la carte la plus petite et on l'insère dans la partie des cartes triées en dernière position

```python
def tri_selection(T, verbose=False):
    '''
    Entrées :
        T est un tableau d'entiers
        verbose par défaut vaut False. Si verbose vaut True, on affiche toutes les étapes
    Sortie : le tableau T avec les nombres triés dans l'ordre croissant
    '''
    
    # Avant de commencer le tri du tableau
    if verbose== True:
            print(T)
    
    # On commence le tri
    for i in range(0,n) :
        indice_min = i
        # on cherche le minimum à partir de T[i]
        for j in range(i+1,n) :
            if T[indice_min] > T[j] :
                indice_min = j
        #on les échange
        T[indice_min], T[i] = T[i], T[indice_min]
        if verbose== True:
                print(T)

    return T
```

```python
# taille du tableau
n = 20

# création par compréhension
T=[randint (0 ,100) for i in range(n)]

tri_selection(T, True)
```
