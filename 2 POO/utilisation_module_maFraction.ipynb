# Utilisation du module `maFraction`


Pour commencer, on importe la classe `Fraction` du module `maFraction`.

**Attention !**, le fichier `maFraction.py` doit se situer au même niveau que le programme qui va l'utiliser.

```python
from maFraction import Fraction
```

Nous pouvons maintenant créer des fractions.

```python
A = Fraction(1,3)
B = Fraction(2,-4)
```

On peut afficher les fractions.

```python
print(A)
```

Où même intérroger sur le type de l'objet

```python
B
```

Vous pouvez enfin faire toutes les opérations voulues

```python
C = A + B
C
```

# Etudions la supériorité de notre objet `Fraction` sur l'objet `float`.

Nous savons tous ce qui va se passer si nous comparons `0.1**2` et `0.01` alors que nous savons bien qu'en math $0,1^2 = 0,01$

```python
0.1**2 == 0.001
```

Mais avec notre objet `Fraction` :

```python
A = Fraction(1,10)
B = Fraction(1, 100)
A**2 == B
```
