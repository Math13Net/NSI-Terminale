# Utilisation du module `maFraction`


Pour commencer, on importe la classe `Fraction` du module `maFraction`.

**Attention !**, le fichier `maFraction.py` doit se situer au même niveau que le programme qui va l'utiliser.

```python
from maFraction import Fraction
```

```python
A = Fraction(1,3)
B = Fraction(2,-4)
```

```python
print(A)
```
