On importe le module `Image` de `PIL`

```python
from PIL import Image
```

On crée une fonction qui remplace un carré de l'image par un carré de couleur uniforme correspondant à la moyenne des couleurs

```python
def carre(nomimage,x,y,cote):
    '''
    nomimage est le nom de l'image choisie
    x et y sont les coordonnées entières du coin en haut à gauche du carré
    cote est la longueure entière du carré
    ---
    cette fonction pourrait être réécrit plus tard 
    dans une version différente pour gérer le débordement
    '''
    img = Image.open('images/'+nomimage)
    largeur, hauteur = img.size
    
    #calcul de la couleur moyenne
    Rsom, Vsom, Bsom = 0, 0, 0
    nb_pixels = 0
    for i in range(x,x+cote):
        for j in range (y,y+cote):
            #il ne faut pas déborder
            if i<largeur and j<hauteur :
                R, V, B = img.getpixel((i,j))
                Rsom += R
                Vsom += V
                Bsom += B
                nb_pixels += 1
    Rmoy = Rsom // nb_pixels
    Vmoy = Vsom // nb_pixels
    Bmoy = Bsom // nb_pixels
    
    #remplissage du carré avec la couleur moyenne
    for i in range(x,x+cote):
        for j in range (y,y+cote):
            #il ne faut pas déborder
            if i<largeur and j<hauteur :
                img.putpixel((i,j), (Rmoy, Vmoy, Bmoy))
    display(img)
```

On teste la fonction.

```python
carre('funnyanimalcat.jpeg', 60, 175, 50)
```

On effectue la mosaïque en utilisant la fonction `carre`.

```python
def carre2(img, x, y, cote):
    '''
    C'est la version à intégrer dans la fonction mosaique
    '''
    largeur, hauteur = img.size
    
    #calcul de la couleur moyenne
    Rsom, Vsom, Bsom = 0, 0, 0
    nb_pixels = 0
    for i in range(x,x+cote):
        for j in range (y,y+cote):
            #il ne faut pas déborder
            if i<largeur and j<hauteur :
                R, V, B = img.getpixel((i,j))
                Rsom += R
                Vsom += V
                Bsom += B
                nb_pixels += 1
    Rmoy = Rsom // nb_pixels
    Vmoy = Vsom // nb_pixels
    Bmoy = Bsom // nb_pixels
    
    #remplissage du carré avec la couleur moyenne
    for i in range(x,x+cote):
        for j in range (y,y+cote):
            #il ne faut pas déborder
            if i<largeur and j<hauteur :
                img.putpixel((i,j), (Rmoy, Vmoy, Bmoy))

def mosaique(nomimage, cote):
    '''
    nom_image est le nom de l'image complet (avec son extension)
    cote est la taille en pixel du coté du carré de la mosaïque
    '''
    img = Image.open('images/'+nomimage)
    largeur, hauteur = img.size
    for x in range(0,largeur,cote):
        for y in range(0,hauteur,cote):
            carre2(img, x, y, cote)
    display(img)
```
