On crée une fonction qui remplace un carré de l'image par un carré de couleur uniforme correspondant à la moyenne des couleurs

```python
def carre(nom_image,x,y,cote):
    img = Image.open('images/'+mon_image)
    
    #calcul de la couleur moyenne
    Rsom, Vsom, Bsom = 0, 0, 0
    for x in range(x,x+cote):
        for y in range (y,y+cote):
            R, V, B = img.getpixel((x,y))
            Rsom += R
            Vsom += V
            Bsom += B
    nb_pixels = cote**2
    Rmoy = Rsom / nb_pixels
    Vmoy = Vsom / nb_pixels
    Bmoy = Bsom / nb_pixels
    
    #remplissage du carré avec la couleur moyenne
    for x in range(x,x+cote):
        for y in range (y,y+cote):
            img.putpixel((x,y), (Rmoy, Vmoy, Bmoy))
    display(img)
         
```
