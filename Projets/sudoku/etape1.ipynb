# Première étape

*Le but de cette première étape est de créer et d'afficher une grille de sudoku dans la console.*
 

On crée la grille sous la forme d'un tableau de tableaux, autrement dit un tableau 2D.

Les cases vides contiennent la valeur `None`.

```python
grille =[
    [None,6,8,1,9,None,5,2,3],
    [5,3,1,8,None,6,9,7,4],
    [9,2,4,7,5,3,8,6,1],
    [2,4,5,None,7,1,3,8,6],
    [3,None,6,2,4,8,1,None,9],
    [1,None,9,6,None,5,7,4,2],
    [8,9,3,5,None,2,4,1,None],
    [6,None,7,4,1,9,None,3,8],
    [4,None,2,3,8,7,6,9,5]]
```

On affiche la grille dans la console.

```python
def affiche(sud):
    
    #3 premières lignes
    for i in range(9):
        if i%3==0:
            print('-'*31)
        ligne =''
        for j in range(9):
            if j%3==0:
                ligne += '|'
            if sud[i][j]==None:
                ligne += '   '
            else :
                ligne += ' '+str(sud[i][j])+' '
        ligne += '|'
        print(ligne)
    print('-'*31)
```

```python
affiche(grille)
```
