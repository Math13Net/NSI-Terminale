# Troisième étape

*Le but de cette troisième étape est de compléter la grille sur les lignes, les colonnes et les carrés pour lesquels il ne manque qu'une seule case.*
 

On va travailler sur la grille de l'étape 1.

On reprend les fonctions `ligne()`, `colonne()` et `carre()` de l'étape 2 qui sont présentes dans le fichier `module_sudoku.py`.

```python
import module_sudoku as mds

grille =[
    [None,6,8,1,9,None,5,2,3],
    [5,3,1,8,None,6,9,7,4],
    [9,2,4,7,5,3,8,6,1],
    [2,4,5,None,7,1,3,8,6],
    [3,None,6,2,4,8,1,None,9],
    [1,None,9,6,None,5,7,4,2],
    [8,9,3,5,None,2,4,1,None],
    [6,None,7,4,1,9,None,3,8],
    [4,None,2,3,8,7,6,9,5]]
```

```python
mds.affiche(grille)
```

On effectue la résolution. Comme c'est une grille de niveau très facile, dans un premier temps, on part du principe que l'on va réussir à résoudre la grille avec cette méthode.

```python
def resolution(sud):
    #on compte le nombre de cases vides au départ
    nb_vides = 0
    for i in range(9):
        for j in range(9):
            if sud[i][j] == None:
                nb_vides += 1
    
    #on affiche la grille
    mds.affiche(sud)
    print("il y a",nb_vides,"cases vides à compléter")
    
    #on démarre la résolution
    while nb_vides != 0:
        #on regarde toutes les lignes
        for i in range(9):
            nb_cases, valeurs = mds.ligne(sud,i)
            if nb_cases == 1:
                #il n'y a qu'une seule case à compléter
                for j in range(9):
                    if sud[i][j] == None:
                        sud[i][j] = valeurs[0]
                        nb_vides -= 1
                        mds.affiche(sud)
                        print("sur la ligne",i,"on vient de rajouter le nombre",valeurs[0])
                        print("il y a",nb_vides,"cases vides à compléter")
        #on regarde toutes les colonnes
        for j in range(9):
            nb_cases, valeurs = mds.colonne(sud,j)
            if nb_cases == 1:
                #il n'y a qu'une seule case à compléter
                for i in range(9):
                    if sud[i][j] == None:
                        sud[i][j] = valeurs[0]
                        nb_vides -= 1
                        mds.affiche(sud)
                        print("sur la colonne",j,"on vient de rajouter le nombre",valeurs[0])
                        print("il y a",nb_vides,"cases vides à compléter")
        #il reste à faire la même chose sur les carrés
            
```

```python
resolution(grille)
```
