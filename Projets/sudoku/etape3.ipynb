On effectue la résolution. Comme c'est une grille de niveau très facile, dans un premier temps, on part du principe que l'on va réussir à résoudre la grille avec cette méthode.

```python
def resolution_niv1(sud, verbose=False):
    '''
    La fonction resolution_niv1 permet de résoudre des sudokus de niveau 1 en regardant uniquement
    les lignes, colonnes ou carrés pour lesquels il manque une case.
    Entrées :
    sud un tableau 2d d'entiers entre 0 et 9 ou de None correspondant à une grille de sudoku
    verbose un booléen.
    S'il est vrai, la résolution se fait avec l'explication à chaque étape de ce qui a été rajouté.
    S'il est faux, il n'y a aucune explication.
    Par défaut, verbose est faux.
    '''
    #on compte le nombre de cases vides au départ
    nb_vides = 0
    for i in range(9):
        for j in range(9):
            if sud[i][j] == None:
                nb_vides += 1
    
    #on affiche la grille
    mds.affiche(sud)
    if verbose :
        print("il y a",nb_vides,"cases vides à compléter")
    
    #on démarre la résolution
    while nb_vides != 0:
        #on regarde toutes les lignes
        for i in range(9):
            nb_cases, valeurs = mds.ligne(sud,i)
            if nb_cases == 1:
                #il n'y a qu'une seule case à compléter et il faut la retrouver dans la ligne
                for j in range(9):
                    if sud[i][j] == None:
                        sud[i][j] = valeurs[0]
                        nb_vides -= 1
                        mds.affiche(sud)
                        if verbose :
                            print("sur la ligne",i,"on vient de rajouter le nombre",valeurs[0])
                            print("il y a",nb_vides,"cases vides à compléter")
        
        #on regarde toutes les colonnes
        for j in range(9):
            nb_cases, valeurs = mds.colonne(sud,j)
            if nb_cases == 1:
                #il n'y a qu'une seule case à compléter et il faut la retrouver dans la colonne
                for i in range(9):
                    if sud[i][j] == None:
                        sud[i][j] = valeurs[0]
                        nb_vides -= 1
                        mds.affiche(sud)
                        if verbose :
                            print("sur la colonne",j,"on vient de rajouter le nombre",valeurs[0])
                            print("il y a",nb_vides,"cases vides à compléter")
        
        #on regarde tous les carrés
        for k in range(9):
            nb_cases, valeurs = mds.carre(sud,k)
            if nb_cases == 1:
                #il n'y a qu'une seule case à compléter et i lfaut la retrouver dans le carré
                #ligne et colonne du carré
                ligne = k//3
                colonne = k%3
                for i in range(3):
                    for j in range(3):
                        if sud[ligne*3+i][colonne*3+j] == None :
                            sud[ligne*3+i][colonne*3+j] = valeurs[0]
                            nb_vides -= 1
                mds.affiche(sud)
                if verbose :
                    print("sur le carré ",k,"on vient de rajouter le nombre",valeurs[0])
                    print("il y a",nb_vides,"cases vides à compléter")
            
```
